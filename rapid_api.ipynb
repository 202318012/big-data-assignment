```python
# importing libraries

import pandas as pd
import requests
```

```python


with open('RapidAPIKey.txt', 'r') as f:
    key = f.read()

# assigning url and headers with api key

url = "https://wft-geo-db.p.rapidapi.com/v1/geo/adminDivisions"

headers = {
	"X-RapidAPI-Key": key,
	"X-RapidAPI-Host": "wft-geo-db.p.rapidapi.com"
}

response = requests.get(url, headers=headers)
```

```python
# getting data

data = response.json()
data = data['data']
data
```

```python

# getting data from another url

url = "https://wft-geo-db.p.rapidapi.com/v1/geo/places"

headers = {
	"X-RapidAPI-Key": key,
	"X-RapidAPI-Host": "wft-geo-db.p.rapidapi.com"
}

response = requests.get(url, headers=headers)

data3 = response.json()
```

```python
data3 = data3['data']
data3
```

```python
# assigning data to respective dataframes

df = pd.DataFrame(data)
df2 = pd.DataFrame(data3)
```

```python
# merging data

merged_df = pd.concat([df,df2])
merged_df = merged_df.reset_index()
merged_df = merged_df.drop(columns = 'index')
merged_df
```

```python
# saving data

merged_df.to_csv('Rapid.csv', index=False)
```
