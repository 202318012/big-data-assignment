```python
# importing libraries

import requests
import pandas as pd
import lxml
from bs4 import BeautifulSoup
```

```python
# assigning headers and getting response

headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:80.0) Gecko/20100101 Firefox/80.0'}
response = requests.get('https://www.amazon.in/s?k=watches+for+men&crid=1QJZEK87WAXTB', headers=headers)
```

```python
response
```

```python
# printing content

html_content = response.content
html_content
```

```python
# saving data in html and text file

f = open("watches.html", "wb")
f.write(html_content)
f.close()

f = open("watches.txt", "wb")
f.write(html_content)
f.close()
```

```python
# parsing data from file using html parser

with open('watches.html', 'rb') as f:
    html_content = f.read()
    soup = BeautifulSoup(html_content, 'lxml')

html_content

# making object of BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')
soup
```

```python
# extracting product details using corresponding html tags and attributes

prod_names = soup.find_all('span', class_ = 'a-size-base-plus a-color-base a-text-normal')
brand_names = soup.find_all('span', class_ = 'a-size-base-plus a-color-base')

stars = soup.find_all('i', class_ = 'a-icon-star-small')

rate_cnt = soup.find_all('span', class_ = 'a-size-base s-underline-text')

price = soup.find_all('span', class_ = 'a-price-whole')
price = price[3:]

org_price = soup.find_all('span', class_ = 'a-offscreen')

bought_cnt = soup.find_all('span', class_ = 'a-size-base a-color-secondary')

disc_rate = soup.find_all('span', class_='a-letter-space') 

org_price = soup.find_all('span', class_ = 'a-price a-text-price')
```

```python
# making list of products along with features

products = [ p.text for p in prod_names]
brands = [b.text for b in brand_names]
stars = [ s.contents[0].text.split(' ')[0] for s in stars]
org_list = [int(o.contents[0].text[1:].replace(',', '')) for o in org_price]
price_list = [ int(p.text[:-1].replace(',', '')) for p in price]

bought_cnt_list = []
disc_list = []
rate_cnts = [int(r.text.replace(',', '')) for r in rate_cnt]

for i in disc_rate:
    x = (i.findNextSiblings())
    x = x[0].text.split('%')[0][1:]
    x = int(x)
    disc_list.append(x)

for b in bought_cnt:
    b = b.text
    if '+' in b:
        b = (b.split('+')[0]) + '+'
        bought_cnt_list.append(b)
```

```python
# assigning details in dataframe

df = pd.DataFrame({'product': products, 'brand': brands, 'price_in_inr': price_list, 'star_ratings': stars, 'rating_count': rate_cnts})
df2 = pd.concat([pd.DataFrame({'original_price' : org_list}), pd.DataFrame({'discount_rate' : disc_list}),  pd.DataFrame({'purchase_count': bought_cnt_list})], axis = 1)

final_df = pd.concat([df, df2], axis=1)
final_df
```

```python
# saving data into csv file

final_df.to_csv('watches.csv', index=False)
```
